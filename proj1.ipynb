```python
import cv2
import mediapipe as mp
import numpy as np
import imageio
import math
import os

# ÿ•ÿπÿØÿßÿØ MediaPipe
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
mp_face = mp.solutions.face_mesh

# ÿßŸÑŸÖÿ≥ÿßÿ±ÿßÿ™ ÿßŸÑÿµÿ≠Ÿäÿ≠ÿ© ŸÑŸÑÿßÿ≥ÿ™ŸäŸÉŸÄÿ±ÿßÿ™
stickers = {
    "shh": r"E:\computer vision\stickers\STK-20250409-WA0035.webp",      # ‚òùÔ∏è ÿ•ÿµÿ®ÿπ ÿπŸÑŸâ ÿßŸÑŸÅŸÖ
    "thumb_up": r"E:\computer vision\stickers\STK-20250409-WA0001.webp",  # üëç ÿ•ÿ®ŸáÿßŸÖ ŸÖÿ±ŸÅŸàÿπ
    "peace": r"E:\computer vision\stickers\STK-20250408-WA0059.webp",     # üôå ÿßŸÑŸäÿØŸäŸÜ ŸÖÿ±ŸÅŸàÿπÿ™ŸäŸÜ
    "shrug": r"E:\computer vision\stickers\STK-20250207-WA0001.webp",     # ü§∑‚Äç‚ôÇÔ∏è ÿßŸÑÿ™ÿ≠Ÿäÿ©
    "happy": r"E:\computer vision\stickers\STK-20241205-WA0004.webp",     # üòÆ ŸÅÿ™ÿ≠ ÿßŸÑŸÅŸÖ
    "nope": r"E:\computer vision\stickers\STK-20241126-WA0032.webp",      # ‚úã ÿßŸÑÿ•ŸäÿØ ŸÇÿØÿßŸÖ
    "finger_up": r"E:\computer vision\stickers\STK-20240725-WA0009.webp"  # ü§¶‚Äç‚ôÇÔ∏è ÿµÿ®ÿßÿπŸäŸÜ ÿπŸÑŸâ ÿßŸÑÿ±ÿ£ÿ≥
}

# ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿßÿ≥ÿ™ŸäŸÉÿ±ÿßÿ™
loaded_stickers = {}
for name, path in stickers.items():
    if os.path.exists(path):
        try:
            loaded_stickers[name] = imageio.mimread(path)
        except:
            loaded_stickers[name] = [cv2.imread(path, cv2.IMREAD_UNCHANGED)]


# ÿØÿßŸÑÿ© ŸÑÿ≠ÿ≥ÿßÿ® ÿßŸÑŸÖÿ≥ÿßŸÅÿ© ÿ®ŸäŸÜ ŸÜŸÇÿ∑ÿ™ŸäŸÜ
def dist(p1, p2):
    return math.sqrt((p1.x - p2.x)**2 + (p1.y - p2.y)**2)


# ÿØÿßŸÑÿ© ŸÑÿ™ÿ≠ÿØŸäÿØ ŸÜŸàÿπ ÿßŸÑÿ•ÿ¥ÿßÿ±ÿ© ÿ£Ÿà ÿßŸÑÿ≠ÿßŸÑÿ©
def detect_gesture(hand_landmarks, face_landmarks, num_hands):
    if not hand_landmarks:
        return "none"

    # ÿ±ÿ§Ÿàÿ≥ ÿßŸÑÿ£ÿµÿßÿ®ÿπ
    tips = [8, 12, 16, 20]
    thumb_tip = 4
    wrist = hand_landmarks.landmark[0]
    index_tip = hand_landmarks.landmark[8]

    # ÿ™ÿ≠ÿØŸäÿØ ÿßŸÑÿ£ÿµÿßÿ®ÿπ ÿßŸÑŸÖÿ±ŸÅŸàÿπÿ©
    fingers = [1 if hand_landmarks.landmark[tip].y < hand_landmarks.landmark[tip - 2].y else 0 for tip in tips]
    thumb_up = hand_landmarks.landmark[thumb_tip].y < hand_landmarks.landmark[thumb_tip - 2].y

    # --- ÿßŸÑÿ™ÿ≠ŸÇŸÇ ŸÖŸÜ ŸÅÿ™ÿ≠ ÿßŸÑŸÅŸÖ ---
    mouth_open = False
    if face_landmarks:
        top_lip = face_landmarks.landmark[13]
        bottom_lip = face_landmarks.landmark[14]
        lip_distance = abs(top_lip.y - bottom_lip.y)
        mouth_open = lip_distance > 0.015  # üëà ŸÅÿ™ÿ≠ ÿßŸÑŸÅŸÖ ÿ®ÿ¥ŸÉŸÑ ŸÖÿ±ŸÜ

    # =================== ÿßŸÑÿ≠ÿ±ŸÉÿßÿ™ ===================

    # ü§´ ÿ•ÿµÿ®ÿπ ÿπŸÑŸâ ÿßŸÑŸÅŸÖ
    if fingers == [1, 0, 0, 0]:
        if face_landmarks:
            nose_tip = face_landmarks.landmark[1]
            if dist(index_tip, nose_tip) < 0.1:
                return "shh"
        return "finger_up"

    # üëç ÿ•ÿ®ŸáÿßŸÖ ŸÖÿ±ŸÅŸàÿπ
    if thumb_up and sum(fingers) == 0:
        return "thumb_up"

    # üôå ÿßŸÑŸäÿØŸäŸÜ ŸÖÿ±ŸÅŸàÿπÿ™ŸäŸÜ
    if num_hands == 2:
        return "peace"

    # ü´° ÿßŸÑÿ™ÿ≠Ÿäÿ© (ÿ•ŸäÿØ ÿπŸÜÿØ ÿßŸÑÿ±ÿ£ÿ≥)
    # ü´° ÿßŸÑÿ™ÿ≠Ÿäÿ© (ÿßŸÑÿ•ŸäÿØ ŸÇÿ±Ÿäÿ®ÿ© ŸÖŸÜ ÿßŸÑÿ±ÿ£ÿ≥ ÿ£Ÿà ÿßŸÑÿπŸäŸÜ)
    if face_landmarks:
    # ŸÜÿ¨Ÿäÿ® ŸÖŸàŸÇÿπ ÿßŸÑÿπŸäŸÜ ÿßŸÑŸäŸÖŸÜŸâ ŸÉŸÖÿ§ÿ¥ÿ± ŸÑŸÑÿ±ÿ£ÿ≥
        right_eye = face_landmarks.landmark[159]  # Ÿàÿßÿ≠ÿØÿ© ŸÖŸÜ ŸÜŸÇÿßÿ∑ ÿßŸÑÿπŸäŸÜ
        head_level = right_eye.y

    # ŸÑŸà ÿßŸÑŸÖÿπÿµŸÖ ÿ£ÿπŸÑŸâ ŸÖŸÜ ŸÖÿ≥ÿ™ŸàŸâ ÿßŸÑÿπŸäŸÜ ÿ£Ÿà ŸÇÿ±Ÿäÿ® ÿ¨ÿØŸãÿß ŸÖŸÜŸáÿß
        if wrist.y < head_level + 0.05:
            return "shrug"


    # üòÆ ŸÅÿ™ÿ≠ ÿßŸÑŸÅŸÖ
    if mouth_open:
        return "happy"

    # ‚úã ÿßŸÑÿ•ŸäÿØ ŸÇÿØÿßŸÖ
    if sum(fingers) >= 3:
        return "nope"

    # ü§¶‚Äç‚ôÇÔ∏è ÿµÿ®ÿßÿπŸäŸÜ ÿπŸÑŸâ ÿßŸÑÿ±ÿ£ÿ≥
    if fingers == [1, 1, 0, 0]:
        return "finger_up"

    return "none"


# ÿ™ÿ¥ÿ∫ŸäŸÑ ÿßŸÑŸÉÿßŸÖŸäÿ±ÿß
cap = cv2.VideoCapture(0)
frame_index = 0

with mp_hands.Hands(min_detection_confidence=0.6, min_tracking_confidence=0.6, max_num_hands=2) as hands, \
     mp_face.FaceMesh(min_detection_confidence=0.6, min_tracking_confidence=0.6) as face_mesh:

    cv2.namedWindow("Camera Feed")
    cv2.namedWindow("Sticker Window")
    cv2.moveWindow("Sticker Window", 700, 100)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame = cv2.flip(frame, 1)
        rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        hand_results = hands.process(rgb)
        face_results = face_mesh.process(rgb)

        gesture = "none"
        num_hands = len(hand_results.multi_hand_landmarks) if hand_results.multi_hand_landmarks else 0

        if hand_results.multi_hand_landmarks:
            for hand_landmarks in hand_results.multi_hand_landmarks:
                mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
                gesture = detect_gesture(
                    hand_landmarks,
                    face_results.multi_face_landmarks[0] if face_results.multi_face_landmarks else None,
                    num_hands
                )

        # ==== ÿπÿ±ÿ∂ ÿßÿ≥ŸÖ ÿßŸÑÿ≠ÿ±ŸÉÿ© ÿπŸÑŸâ ÿßŸÑŸÉÿßŸÖŸäÿ±ÿß ====
        cv2.putText(frame, f"Gesture: {gesture}", (20, 40),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        # ==== ÿπÿ±ÿ∂ ÿßŸÑŸÉÿßŸÖŸäÿ±ÿß ====
        cv2.imshow("Camera Feed", frame)

        # ==== ÿπÿ±ÿ∂ ÿßŸÑÿßÿ≥ÿ™ŸäŸÉÿ± ÿßŸÑŸÖŸÜÿßÿ≥ÿ® ====
        if gesture in loaded_stickers:
            frames = loaded_stickers[gesture]
            frame_index += 1
            sticker_frame = frames[frame_index % len(frames)]

            # ÿ™ÿ≠ŸàŸäŸÑ ÿµŸäÿ∫ÿ© ÿßŸÑŸÑŸàŸÜ
            if sticker_frame.shape[2] == 4:
                sticker_img = cv2.cvtColor(np.array(sticker_frame), cv2.COLOR_RGBA2BGR)
            else:
                sticker_img = cv2.cvtColor(np.array(sticker_frame), cv2.COLOR_RGB2BGR)

            # ÿ™ÿ∫ŸäŸäÿ± ÿßŸÑÿ≠ÿ¨ŸÖ
            sticker_resized = cv2.resize(sticker_img, (350, 350))
            cv2.imshow("Sticker Window", sticker_resized)
        else:
            blank = np.zeros((350, 350, 3), dtype=np.uint8)
            cv2.putText(blank, "No gesture", (50, 180),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.2, (200, 200, 200), 2)
            cv2.imshow("Sticker Window", blank)

        # ÿßŸÑÿÆÿ±Ÿàÿ¨ ÿ®ÿßŸÑÿ∂ÿ∫ÿ∑ ÿπŸÑŸâ Q
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
